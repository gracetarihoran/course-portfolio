### Data Cleaning

Based on the initial data overview, we have to do some cleaning process in the raw data:

1. **Handled missing values**
     - Filled missing values in column `Gender` with `Unknown` 
     - Filled missing values in column `Discount value` with `0`
       
2. **Handled invalid value**
     - Remove the rows with values `0` in column `Age`
      
3. **Handle inconsistent values**
    - Replace value Female to F, Male to M, and Unknown to U (standardized) in column `Gender`
    - Replace value `nothing` to `0` in column `Previous Purchases`
    - Replace values `Debit` to `Debit Card` in column `Payment Method`
      
4. **Change column name**
    - Change the column name `Discount value` to `Discount Value (%)` to make it clear that the values are in %
      
6. **Remove unnecessary columns**
    - Column `Year` and `Country` were removed because they contained only one unique value and therefore did not provide useful information for the analysis

```python
# Cleaning
# Impute missing values in the Gender column with "Unknown"
df['Gender'] = df['Gender'].fillna('Unknown')

# Impute '0' to missing values in column 'Discount value'
df['Discount value'] = df['Discount value'].fillna(0)

# Remove rows where Age is 0 because 0 is considered an invalid age value
df = df[df['Age'] != 0]

# Standardize Gender inconsistent values
# Change Female to F, Male to M, and Unknown to U
df['Gender'] = df['Gender'].replace({
    'Female': 'F',
    'Male': 'M',
    'Unknown': 'U'
})

# Standardize values in the Previous Purchases column
# Clean text format and replace "nothing" with 0
df['Previous Purchases'] = df['Previous Purchases'].replace('nothing', 0).astype(int)

# Standardize values in the Payment Method column
# Change Debit" to "Debit Card"
df['Payment Method'] = df['Payment Method'].replace({
    'Debit': 'Debit Card'
})

# Change the column name Discount value to Discount Value (%)
df = df.rename(columns={
    "Discount value": "Discount Value (%)"
})

# Remove unnecessary columns
df = df.drop(columns=["Year", "Country"])

# Check the number of rows and columns in the dataset
print(f"Dataset Shape: {df.shape}")

# Save cleaned dataset
df.to_csv("shopping_data_cleaned.csv", index=False)
print("\nData cleaning completed successfully.")
print("The cleaned dataset has been saved as 'shopping_data_cleaned.csv'.")
```

### Task 1. Analyze whether promo codes increase customer purchase amounts across different age groups

The essential goal of this task is growth validation. This task helps evaluate whether promo codes are actually related to higher customer spending or whether the business is only giving discounts without gaining additional purchase value.

This task focuses on two main questions:
1. Do customers who use promo codes spend more than customers who do not use promo codes?
2. Do different age groups respond to promo codes in the same way?

```python
# Define descriptive age groups
bins = [0, 18, 30, 50, 65, float("inf")]
labels = ["Teens", "Young Adults", "Adults", "Seniors", "Elderly"]

df["Age Group"] = pd.cut(
    df["Age"],
    bins=bins,
    labels=labels,
    right=True,
    include_lowest=True
)

# Define custom colors for the plots
# Green for 'Yes' (Promo Used), Orange for 'No' (No Promo)
custom_palette = {
    "Yes": "#2E8B57",
    "No": "#FF8C00"
}

# Create the Bar Chart: Average Purchase Amount
summary = df.groupby(
    ["Age Group", "Promo Code Used"],
    observed=True
)["Purchase Amount (USD)"].mean().reset_index()

plt.figure(figsize=(12, 6))

sns.barplot(
    data=summary,
    x="Age Group",
    y="Purchase Amount (USD)",
    hue="Promo Code Used",
    order=labels,
    hue_order=["Yes", "No"],
    palette=custom_palette,
)

plt.title("Task 1.1: Average Purchase Amount by Age Group and Promo Usage")
plt.xlabel("Age Group")
plt.ylabel("Mean Purchase Amount (USD)")
plt.legend(title="Promo Code Used")
plt.tight_layout()
plt.show()


# Create the Boxplot: Purchase Amount Distribution
plt.figure(figsize=(12, 6))

sns.boxplot(
    data=df,
    x="Age Group",
    y="Purchase Amount (USD)",
    hue="Promo Code Used",
    order=labels,
    hue_order=["Yes", "No"],
    palette=custom_palette
)

plt.title("Task 1.2: Purchase Amount Distribution by Age Group and Promo Usage")
plt.xlabel("Age Group")
plt.ylabel("Purchase Amount (USD)")
plt.legend(title="Promo Code Used")
plt.tight_layout()
plt.show()
```

**Interpretation:**

The two charts were used to answer the task: Analyze whether promo codes are associated with higher customer purchase amounts across different age groups.

**1.1 Bar chart:**

- The bar chart compares the average purchase amount between customers who used promo codes and customers who did not, across different age groups.

- For Young Adults and Adults, the green bars are slightly higher than the orange bars. This suggests that customers in these age groups who used promo codes had a higher average purchase amount compared to those who did not use promo codes.

- For Teens, the orange bar is higher than the green bar. This suggests that promo code usage was not associated with higher average spending in this group. However, the number of teen customers is relatively small, so this result should be interpreted carefully.

- For Seniors and Elderly customers, the difference between promo-code users and non-promo users is small. This suggests that promo codes may not have a strong relationship with the average purchase amount in these age groups.
  
**1.2 Boxplot:**

- The boxplot shows the distribution of purchase amounts for each age group and promo code status. This is important because the bar chart only shows the average, while the boxplot shows the median, spread, and overall range of spending.

- The median purchase amount is quite similar across most age groups, regardless of whether customers used promo codes or not. This suggests that the typical customer spending level did not change strongly with promo code usage.

- The boxplot also shows that both promo-code users and non-promo users still made purchases across a wide range, from lower-value purchases to higher-value purchases. This means promo-code users were not only buying low-value products. They also continued to make higher-value purchases.

- Overall, the boxplot supports the bar chart by showing that the small differences in average purchase amount are not caused only by one or two extreme purchases. However, the overall spending distribution remains fairly similar between promo-code users and non-promo users.

**Conclusion:**

- The visual analysis suggests that promo codes are most positively associated with higher average purchase amounts among Young Adults and Adults. However, the effect is not equally visible across all age groups. For Seniors and Elderly customers, the difference is small, while for Teens, promo-code users do not appear to spend more on average.

- Therefore, promo codes may be more useful when targeted toward Young Adults and Adults. At the same time, the boxplot suggests that promo-code users are not only using discounts for cheap purchases, because they still purchase across the full spending range, including higher-value purchases.

### Task 2. Identify states with high sales volume but low customer review ratings

For this task, our main focus is risk management and profit protection. High sales volume is good for the business, but if those same states also have low review ratings, it means the business may be earning revenue while customers are not fully satisfied.

This is risky because unhappy customers may stop buying or move to competitors. Therefore, states with high sales but low ratings should be treated as priority risk areas.

By identifying these states, the business can focus its resources on the most important problems first. After that, managers can investigate the causes of low ratings and decide what improvements are needed.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Aggregate the data by State (Location)
state_data = (
    df.groupby('Location')
    .agg({
        'Purchase Amount (USD)': 'sum',
        'Review Rating': 'mean'
    })
    .reset_index()
)

# Rename columns for clearer chart labels
state_data = state_data.rename(columns={
    'Purchase Amount (USD)': 'Total Sales',
    'Review Rating': 'Average Review Rating'
})

# Set thresholds using the average of all states
sales_threshold = state_data['Total Sales'].mean()
rating_threshold = state_data['Average Review Rating'].mean()

# Create a risk group column
state_data['Risk Group'] = 'Other states'

state_data.loc[
    (state_data['Total Sales'] > sales_threshold) &
    (state_data['Average Review Rating'] < rating_threshold),
    'Risk Group'
] = 'High sales, lower rating'

# Assign colors based on risk group
colors = state_data['Risk Group'].map({
    'Other states': 'lightgray',
    'High sales, lower rating': 'seagreen'
})


# Function to draw scatter plot
def draw_scatter_plot(ax, x_limits, title, xlabel, show_labels=False):
    
    ax.scatter(
        state_data['Average Review Rating'],
        state_data['Total Sales'],
        s=90,                 # Fixed size, no bubble encoding
        c=colors,             # Color only highlights risk group
        alpha=0.85,
        edgecolors='black',
        linewidth=0.6
    )

    # Threshold lines
    ax.axvline(
        rating_threshold,
        color='gray',
        linestyle='--',
        linewidth=1
    )

    ax.axhline(
        sales_threshold,
        color='gray',
        linestyle='--',
        linewidth=1
    )

    # Set x-axis range
    ax.set_xlim(x_limits)

    # Titles and labels
    ax.set_title(title, fontsize=13)
    ax.set_xlabel(xlabel)
    ax.set_ylabel('Total Sales (USD)')

    # Grid
    ax.grid(True, alpha=0.3)

    # Label states in the priority risk area only on the zoomed chart
    if show_labels:
        for i in range(state_data.shape[0]):
            row = state_data.iloc[i]

            if row['Risk Group'] == 'High sales, lower rating':
                ax.text(
                    row['Average Review Rating'] + 0.003,
                    row['Total Sales'] + 40,
                    row['Location'],
                    fontsize=9,
                    fontweight='bold'
                )

        # Quadrant labels
        ax.text(
            x_limits[0] + 0.01,
            sales_threshold + 200,
            'High sales\nLower rating',
            fontsize=10,
            fontweight='bold'
        )

        ax.text(
            rating_threshold + 0.01,
            sales_threshold + 200,
            'High sales\nHigher rating',
            fontsize=10,
            fontweight='bold'
        )


# Create two charts side by side
fig, axes = plt.subplots(
    1, 2,
    figsize=(18, 7),
    sharey=True,
    constrained_layout=True
)

# Chart 1: Full rating scale from 1 to 5
draw_scatter_plot(
    ax=axes[0],
    x_limits=(1, 5),
    title='Full Rating Scale View',
    xlabel='Average Review Rating (1-5 scale)',
    show_labels=False
)

# Chart 2: Zoomed view from 3.5 to 4.0
draw_scatter_plot(
    ax=axes[1],
    x_limits=(3.5, 4.0),
    title='Zoomed View of Observed Ratings',
    xlabel='Average Review Rating (zoomed to 3.5-4.0)',
    show_labels=True
)

# Main title
fig.suptitle(
    'Task 2. State Performance: Total Sales vs. Customer Rating',
    fontsize=16,
    fontweight='bold'
)

# Create custom legend
legend_elements = [
    Line2D(
        [0], [0],
        marker='o',
        color='w',
        label='Other states',
        markerfacecolor='lightgray',
        markeredgecolor='black',
        markersize=9
    ),
    Line2D(
        [0], [0],
        marker='o',
        color='w',
        label='High sales, lower rating',
        markerfacecolor='seagreen',
        markeredgecolor='black',
        markersize=9
    ),
    Line2D(
        [0], [0],
        color='gray',
        linestyle='--',
        label='Average threshold'
    )
]

axes[1].legend(
    handles=legend_elements,
    title='State group',
    loc='upper right'
)

# Add note under the charts
fig.text(
    0.5,
    -0.04,
    'Note: Review rating is originally measured on a 1-5 scale. The right chart is zoomed only to make small differences easier to read.',
    ha='center',
    fontsize=9
)

plt.show()


# Filter states in the priority risk area:
# Above-average sales and below-average rating
priority_risk_states = state_data[
    (state_data['Total Sales'] > sales_threshold) &
    (state_data['Average Review Rating'] < rating_threshold)
]

# Sort by total sales
top_5_priority_risk = priority_risk_states.sort_values(
    by='Total Sales',
    ascending=False
).head(5)

print("TOP 5 STATES: HIGH SALES BUT LOWER RATINGS")
print(top_5_priority_risk[['Location', 'Total Sales', 'Average Review Rating']])
```

**Observation:**
- The visualization presents two versions of the same bubble chart. The first chart uses the full original review rating scale from 1.0 to 5.0, while the second chart zooms in to the observed rating range from 3.5 to 4.0.
- The x-axis represents the average review rating. States located further to the right have higher average customer satisfaction. The y-axis represents the total purchase amount, where states placed higher on the chart have higher sales volume.
- Bubble size also represents sales volume, so larger bubbles indicate states with higher total purchase amounts. ### Task 4. Analyze sales performance across product categories and regions

For this task, we want to understand how sales are distributed across product categories and regions. 
- This helps us see whether certain regions prefer certain product categories and which locations contribute the most to sales.
- We use a treemap to show the overall structure of region and product category sales
- We use also bar chart to show the top 5 locations with the highest purchase amount.
- This can help the business make better decisions about inventory, marketing, and regional sales strategy.

```python
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns

# Treemap
import plotly.express as px

# Aggregate sales by state and product category
treemap_data = (
    df.groupby(["Location", "Category"], observed=True)["Purchase Amount (USD)"]
    .sum()
    .reset_index()
    .rename(columns={"Purchase Amount (USD)": "Total Sales"})
)

# Calculate total purchase amount
total_purchase = treemap_data["Total Sales"].sum()

# Custom colors for product category
category_colors = {
    "Clothing": "#5DADE2",
    "Accessories": "#AF7AC5",
    "Footwear": "#58D68D",
    "Outerwear": "#EC7063"
}

# Create treemap
fig = px.treemap(
    treemap_data,
    path=["Location", "Category"],
    values="Total Sales",
    color="Category",
    color_discrete_map=category_colors,
    title=(
        "<b>Task 4.1: Sales Performance by State and Product Category</b><br>"
        f"Total Purchase Amount: ${total_purchase:,.2f}"
    )
)

fig.update_traces(
    sort=True,
    textinfo="label+value",
    hovertemplate=
    "<b>%{label}</b><br>" +
    "Total Sales: $%{value:,.2f}<br>" +
    "<extra></extra>",
    marker=dict(
        line=dict(width=1, color="white")
    )
)

# Add outside border around the plot area
fig.update_layout(
    margin=dict(t=90, l=20, r=20, b=20),
    paper_bgcolor="white",
    plot_bgcolor="white",
    shapes=[
        dict(
            type="rect",
            xref="paper",
            yref="paper",
            x0=0,
            y0=0,
            x1=1,
            y1=1,
            line=dict(
                color="black",
                width=1
            ),
            fillcolor="rgba(0,0,0,0)"
        )
    ]
)

fig.show()

# Grouped Bar Chart Top 5 states by total sales
# Find top 5 states based on total purchase amount
top5_states = (
    df.groupby("Location")["Purchase Amount (USD)"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .index
)

# Filter data for only top 5 states
df_top5 = df[df["Location"].isin(top5_states)]

# Summarize sales by state and product category
sales_summary = (
    df_top5.groupby(["Location", "Category"], observed=True)["Purchase Amount (USD)"]
    .sum()
    .reset_index()
    .rename(columns={"Purchase Amount (USD)": "Total Sales"})
)

# Keep states ordered by total sales
state_order = (
    df_top5.groupby("Location")["Purchase Amount (USD)"]
    .sum()
    .sort_values(ascending=False)
    .index
)

# Calculate total sales for each state
state_totals = (
    df_top5.groupby("Location")["Purchase Amount (USD)"]
    .sum()
    .reindex(state_order)
)

# Custom color palette for product categories
category_palette = {
    "Clothing": "#5DADE2",
    "Accessories": "#AF7AC5",
    "Footwear": "#58D68D",
    "Outerwear": "#EC7063"
}

# Create grouped bar chart
plt.figure(figsize=(12, 6))

ax = sns.barplot(
    data=sales_summary,
    x="Location",
    y="Total Sales",
    hue="Category",
    order=state_order,
    palette=category_palette
)

# Add total sales label above each state group only
for i, state in enumerate(state_order):
    total_value = state_totals.loc[state]
    
    ax.text(
        i,
        sales_summary[sales_summary["Location"] == state]["Total Sales"].max() + 250,
        f"Total: ${total_value:,.0f}",
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold"
    )

plt.title("Task 4.2: Sales Performance Across Product Categories in Top 5 States")
plt.xlabel("State")
plt.ylabel("Total Sales (USD)")

plt.legend(
    title="Product Category",
    bbox_to_anchor=(1.05, 1),
    loc="upper left"
)

# Add extra space at the top for total labels
plt.ylim(0, sales_summary["Total Sales"].max() * 1.3)

plt.tight_layout()
plt.show()
```

**Observation:**

**Treemap**
- The treemap shows sales performance across all states and product categories. The size of each rectangle represents the total purchase amount, so larger rectangles indicate a higher sales contribution. Color is used to distinguish product categories, making it easier to identify which categories contribute most to sales across states.
- Clothing is the strongest product category, with total sales of 96,302 USD, followed by Accessories with 69,007 USD. Footwear contributes 34,202 USD, while Outerwear has the lowest total sales with 17,301 USD. This shows that total sales are mainly driven by Clothing and Accessories.

**Grouped bar chart**
- The grouped bar chart focuses on the top five states with the highest total purchase amount: Idaho, Montana, Illinois, Alabama, and Nevada. The total sales among these states are very close. Idaho has 5,308 USD, Montana has 5,206 USD, Illinois has 5,153 USD, while Alabama and Nevada each have 5,134 USD.
- In these top five states, Clothing is generally the strongest product category. Accessories also contribute strongly, especially in Nevada and Illinois. Footwear and Outerwear tend to contribute less compared with Clothing and Accessories.

**Interpretation:**
- The treemap provides an overall view of how sales are distributed across all states and product categories. It shows that the business is primarily supported by Clothing and Accessories, while Footwear and Outerwear contribute less to sales. This suggests that Clothing and Accessories are the main drivers of product sales in the dataset.
- The grouped bar chart gives a more detailed comparison of the strongest sales locations. Although Idaho has the highest total purchase amount, the difference between the top five states is small. This means that sales performance is relatively balanced among the strongest states, rather than being dominated by only one location.

Together, the treemap and grouped bar chart show both the overall pattern and the detailed comparison.
